```python
# Mengecek jumlah data kosong tiap kolom
df.isnull().sum()
```

```python
# Jika ingin mengisi nilai kosong dengan rata-rata (untuk kolom numerik)
df = df.fillna(df.mean(numeric_only=True))
```

```python
# Jika ingin menghapus baris yang memiliki data kosong:
df = df.dropna()
```

```python
# Mengubah kolom bertipe object/kategori menjadi numerik:
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].astype('category').cat.codes

print("Semua kolom sudah menjadi numerik")
```

```python
# Cek kembali tipe data:
df.dtypes
```

```python
# Normalisasi setiap tabel
df_normalized = (df - df.min()) / (df.max() - df.min())

# Menampilkan hasil normalisasi
df_normalized.head()
```
